```python
#Q1
import random
import string

def generate_random_string(length=10):
    characters = string.ascii_letters + string.digits
    return ''.join(random.choice(characters) for i in range(length))

def create_random_string_file(filename="random_strings.txt", num_lines=1000, string_length=20):
    try:
        with open(filename, 'w') as f:
            for i in range(num_lines):
                random_str = generate_random_string(string_length)
                f.write(random_str + '\n')
        print(f"Successfully created '{filename}' with {num_lines} lines of random strings.")
    except IOError as e:
        print(f"Error writing to file {filename}: {e}")

if __name__ == "__main__":
      create_random_string_file(filename="random_strings.txt", num_lines=1000, string_length=20)
```

```python
#Q5
import random
import string
import os

def generate_random_string(length=10):

    characters = string.ascii_letters + string.digits
    return ''.join(random.choice(characters) for i in range(length))

def create_random_string_file(filename_prefix="random_strings", file_count=1, target_file_size_mb=5, string_length=20):

    target_bytes = target_file_size_mb * 1024 * 1024
    created_filenames = []

    for i in range(1, file_count + 1):

        filename = f"{filename_prefix}_{target_file_size_mb}MB_{i}.txt"
        current_bytes_written = 0

        try:
            with open(filename, 'w') as f:

                while current_bytes_written < target_bytes:
                    random_str = generate_random_string(string_length)
                    line = random_str + '\n'
                    f.write(line)

                    current_bytes_written += len(line.encode('utf-8'))


            actual_file_size = os.path.getsize(filename)
            print(f"Successfully created '{filename}' with an approximate size of {actual_file_size / (1024 * 1024):.2f} MB.")
            print(f"Target size for this file was {target_file_size_mb} MB.")
            created_filenames.append(filename)

        except IOError as e:
            print(f"Error writing to file {filename}: {e}")
        print("-" * 50)
    return created_filenames

def convert_file_to_uppercase(filename):

    temp_filename = filename + ".tmp"
    try:
        with open(filename, 'r') as infile, open(temp_filename, 'w') as outfile:
            for line in infile:
                outfile.write(line.upper())

        os.replace(temp_filename, filename)
        print(f"Successfully converted '{filename}' to uppercase.")
    except FileNotFoundError:
        print(f"Error: File '{filename}' not found.")
    except IOError as e:
        print(f"Error processing file '{filename}': {e}")
    except Exception as e:
        print(f"An unexpected error occurred with file '{filename}': {e}")


if __name__ == "__main__":

    file_sizes_gb = [1, 2, 3, 4, 5]
    all_created_files = []

    for size_gb in file_sizes_gb:

        target_size_mb = size_gb * 1024
        print(f"Attempting to create a {size_gb}GB file (approx {target_size_mb} MB)...")


        created_files_for_size = create_random_string_file(
            filename_prefix="random_strings",
            file_count=1,
            target_file_size_mb=target_size_mb,
            string_length=20
        )
        all_created_files.extend(created_files_for_size)

    print("\n" + "="*50)
    print("Starting conversion of created files to uppercase...")
    print("="*50 + "\n")

    for file_to_convert in all_created_files:
        convert_file_to_uppercase(file_to_convert)
```

```python
#Q12
import pandas as pd
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt

# --- (i) Create a random dataset of 100 rows and 30 columns ---
print("--- Step 1: Creating a random dataset (100 rows, 30 columns) ---")
# Generate random integers between 1 and 200
data = np.random.randint(1, 201, size=(100, 30))
df = pd.DataFrame(data, columns=[f'col_{i}' for i in range(30)])
print("Original DataFrame head:")
print(df.head())
print("\n" + "="*80 + "\n")

# --- (ii) Replace all the values with NA in the dataset defined between [10, 60] ---
print("--- Step 2: Replacing values between [10, 60] with NA ---")
# Create a boolean mask where values are between 10 and 60 (inclusive)
mask = (df >= 10) & (df <= 60)
df_with_na = df.where(~mask) # Replace values where mask is True with NaN (NA)

print("DataFrame head after replacing values with NA:")
print(df_with_na.head())

# Print the count of number rows having missing values
rows_with_missing_values = df_with_na.isnull().any(axis=1).sum()
print(f"\nNumber of rows with missing values (NA): {rows_with_missing_values}")
print("\n" + "="*80 + "\n")

# --- (iii) Replace all the NA values with the average of the column value ---
print("--- Step 3: Replacing NA values with the average of their respective columns ---")
df_filled = df_with_na.fillna(df_with_na.mean())
print("DataFrame head after filling NA values with column averages:")
print(df_filled.head())
print("\n" + "="*80 + "\n")

# --- (iv) Find the Pearson correlation among all the columns and plot heat map ---
print("--- Step 4: Pearson Correlation Heatmap & Column Selection ---")
# Calculate Pearson correlation matrix
correlation_matrix = df_filled.corr(method='pearson')

# Plotting the heatmap
plt.figure(figsize=(16, 12))
sns.heatmap(correlation_matrix, annot=False, cmap='coolwarm', fmt=".2f", linewidths=.5)
plt.title('Pearson Correlation Heatmap of Dataset Columns')
plt.show()

# Select those columns having correlation <= 0.7
# We will identify pairs of columns where the absolute correlation coefficient is <= 0.7
print("\nPairs of columns with absolute Pearson correlation <= 0.7:")
low_correlation_pairs = []
# Iterate through the upper triangle of the correlation matrix to avoid duplicates
for i in range(len(correlation_matrix.columns)):
    for j in range(i + 1, len(correlation_matrix.columns)):
        col1 = correlation_matrix.columns[i]
        col2 = correlation_matrix.columns[j]
        corr_value = correlation_matrix.loc[col1, col2]
        if abs(corr_value) <= 0.7:
            low_correlation_pairs.append((col1, col2, corr_value))

if low_correlation_pairs:
    for pair in low_correlation_pairs:
        print(f"  {pair[0]} - {pair[1]}: Correlation = {pair[2]:.4f}")
else:
    print("  No pairs of columns found with absolute correlation <= 0.7.")
print("\n" + "="*80 + "\n")

# --- (v) Normalize all the values in the dataset between 0 and 10 ---
print("--- Step 5: Normalizing all values in the dataset between 0 and 10 ---")
df_normalized = df_filled.copy() # Work on a copy to preserve previous states

# Min-Max Scaling formula: X_scaled = (X - X_min) / (X_max - X_min) * (new_max - new_min) + new_min
# Here, new_min = 0, new_max = 10
# We apply this column-wise to ensure correct min/max for each column
for column in df_normalized.columns:
    min_val = df_normalized[column].min()
    max_val = df_normalized[column].max()

    # Avoid division by zero if all values in a column are the same
    if (max_val - min_val) != 0:
        df_normalized[column] = ((df_normalized[column] - min_val) / (max_val - min_val)) * 10
    else:
        # If all values are the same, they should all map to the new_min (0 in this case)
        df_normalized[column] = 0

print("DataFrame head after normalization (values between 0 and 10):")
print(df_normalized.head())
print("\n" + "="*80 + "\n")

# --- (vi) Replace all the values in the dataset with 1 if value <= 0.5 else with 0 ---
print("--- Step 6: Replacing values: 1 if <= 0.5, else 0 ---")
# Apply a lambda function across the entire DataFrame
df_binary = df_normalized.applymap(lambda x: 1 if x <= 0.5 else 0)

print("DataFrame head after binary transformation:")
print(df_binary.head())
print("\n" + "="*80 + "\n")

# Optional: Display some statistics of the final binary dataset
print("Value counts for a sample column (e.g., col_0) after binary transformation:")
print(df_binary['col_0'].value_counts())


```

```python
#Q14
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt

# Step 1: Create a random dataset of 600 rows and 15 columns with values between [-100, 100]
np.random.seed(42)
data = np.random.randint(-100, 101, size=(600, 15))
df = pd.DataFrame(data, columns=[f'Col_{i+1}' for i in range(15)])

# (i) Plot scatter graph between Column 5 and Column 6
plt.figure(figsize=(6, 4))
plt.scatter(df['Col_5'], df['Col_6'], color='purple', alpha=0.6)
plt.title('Scatter Plot between Column 5 and Column 6')
plt.xlabel('Column 5')
plt.ylabel('Column 6')
plt.grid(True)
plt.show()

# (ii) Plot histogram of each column in single graph
plt.figure(figsize=(10, 6))
for col in df.columns:
    plt.hist(df[col], bins=30, alpha=0.3, label=col)
plt.title('Histogram of All Columns')
plt.xlabel('Value')
plt.ylabel('Frequency')
plt.legend(loc='upper right', fontsize='small')
plt.show()

# (iii) Plot Box plot of each column in a single graph
plt.figure(figsize=(12, 6))
df.boxplot()
plt.title('Box Plot of All Columns')
plt.xlabel('Columns')
plt.ylabel('Values')
plt.xticks(rotation=45)
plt.grid(True)
plt.show()
```
